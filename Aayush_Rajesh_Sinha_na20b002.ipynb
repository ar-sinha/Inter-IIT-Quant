```python
url = 'https://raw.githubusercontent.com/ar-sinha/Inter-IIT-Quant/main/quant_dataset.csv'
df = pd.read_csv(url)
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df
```

In the cell above:  
1. We read the dataset into a dataframe  
1. Dropping the index column  
1. Convert the data to index based on date

```python
df.interpolate(method='linear', inplace=True)
df
```

In the cell above:
1. We observe that some values in the dataset are missing
1. We linearly interpolate those values

```python
date_range = pd.date_range('2005-01-03','2016-11-30')
df = df.reindex(date_range).interpolate(method='linear')
df.to_csv('cleaned_data.csv', index=True)
df
```

For a dataset to be treated and processed as time series, it should have a constant time period between consecutive observations but we observe that is not the case here.  
Hence, we add the missing dates and again interpolate based on neighbouring datapoints, in the cell above.

```python
import matplotlib.pyplot as plt

for col in df.columns:
    df[col].plot()
    plt.title(col)
    plt.show()
```

Now, we try to identify if any of the given signals is white noise, so that we can eliminate that parameter.  
As a basic check we plot the graphs of signals.
It turns out that none of the signals has a mean value of zero, so we can conclude that none of them is indeed white noise.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
results = seasonal_decompose(df['Price'])
results.plot()
```

Apart from the signals the Price on a certain day shall also depend on price on the previous days.  
To accomodate this, we add new columns (as parameters) which indicate price yesterday, price 2 days ago and price 3 days ago.  

```python
df['day-1'] = df['Price'].shift(+1)
df['day-2'] = df['Price'].shift(+2)
df['day-3'] = df['Price'].shift(+3)
df
```

This operation creates 3 NaN values, which we should drop.

```python
df = df.dropna()
df
```

We try 2 popular regression models, Random Forest Regressor and Linear Regressor

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 100, max_features = 3, random_state = 1)
```

We convert the dataframe into X and Y so that we can use that as input and output in our model.  
X being the input would contain all parameters which includes 7 signals and 3 previous day prices.  
Y would be the actual value of the price on that day.

```python
X = df.iloc[:, 1:].values.tolist()
Y = df['Price'].tolist()
```

We split the data to keep the last 100 datapoints for testing and use the rest for training.  
We fit the Random Forest Regressor model below and graph the test results (predictions) vs actual price.

```python
X_train, X_test, Y_train, Y_test = X[:-100], X[-100:], Y[:-100], Y[-100:]

model.fit(X_train, Y_train)
pred = model.predict(X_test)
plt.rcParams["figure.figsize"] = (12, 8)
plt.plot(pred, label='Random_Forest_Predictions')
plt.plot(Y_test, label='Actual Price')
plt.legend(loc="upper left")
plt.show()
```

Next, we try to fit the Linear Regression model and again we plot the test results (predictions) vs actual price.

```python
lin_model.fit(X_train, Y_train)
lin_pred = lin_model.predict(X_test)
plt.rcParams["figure.figsize"] = (12, 8)
plt.plot(lin_pred, label='Linear_Regression_Predictions')
plt.plot(Y_test, label='Actual Price')
plt.legend(loc="upper left")
plt.show()
```

To compare the accuracy of both the models, we find the mean squared error.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

mse_rf = sqrt(mean_squared_error(pred, Y_test))
mse_lr = sqrt(mean_squared_error(lin_pred, Y_test))

print("Mean squared error for Random Forest Regressor is: ", mse_rf)
print("Mean squared error for Linear Regressor is: ", mse_lr)
```

The linear regression model predicts the price quite accurately and shows 6 times lesser mean squared error.
